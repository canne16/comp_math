```python
import csv
import matplotlib.pyplot as plt

def read_csv(filename):
    """
    Reads data from a CSV file and returns two lists: x and y values.
    """
    x = []
    y = []
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header row
        for row in reader:
            x.append(float(row[0]))
            y.append(float(row[1]))
    return x, y

def plot_functions(x1, y1, x2, y2, x3, y3, labels, title):
    """
    Plots three functions on the same graph.
    """
    plt.figure(figsize=(10, 6))
    plt.plot(x1, y1, label=labels[0], color='blue')
    plt.plot(x2, y2, label=labels[1], color='red')
    plt.plot(x3, y3, label=labels[2], color='green')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    # Read data from CSV files
    data_dir = "../../Cpp/compmath/export_data/"
    
    x_func, y_func = read_csv(data_dir + "function.csv")
    x_der, y_der = read_csv(data_dir + "derivative.csv")
    x_sec_der, y_sec_der = read_csv(data_dir + "sec_der.csv")
    x_true_der, y_true_der = read_csv(data_dir + "true_der.csv")
    x_true_sec_der, y_true_sec_der = read_csv(data_dir + "true_sec_der.csv")

    # Plot the functions
    plot_functions(
        x_func, y_func, 
        x_der, y_der, 
        x_sec_der, y_sec_der,
        labels=["Function", "Derivative", "Second Derivative"], 
        title="Function and Its Derivative"
    )
```

```python
def plot_functions(x1, y1, x2, y2, labels, title):
    """
    Plots two functions on the same graph.
    """
    plt.figure(figsize=(10, 6))
    plt.plot(x1, y1, label=labels[0], color='blue')
    plt.plot(x2, y2, label=labels[1], color='red', linestyle=':')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_functions(
        x_der, y_der, 
        x_true_der, y_true_der, 
        labels=["Derivative", "True Derivative"], 
        title="Derivative and Its Approximation"
    )
```

```python
plot_functions(
        x_sec_der, y_sec_der, 
        x_true_sec_der, y_true_sec_der, 
        labels=["Second Derivative", "True Second Derivative"], 
        title="Second Derivative and Its Approximation"
    )
```
