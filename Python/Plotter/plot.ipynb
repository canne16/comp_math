```python
import csv
import matplotlib.pyplot as plt

def read_csv(filename):
    x = []
    y = []
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header row
        for row in reader:
            x.append(float(row[0]))
            y.append(float(row[1]))
    return x, y

def plot_functions(x, ys, labels, title):
    plt.figure(figsize=(10, 6))
    for y, label in zip(ys, labels):
        plt.plot(x, y, label=label)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_loglog_functions(x, ys, labels, title):
    plt.figure(figsize=(10, 6))
    for y, label in zip(ys, labels):
        plt.loglog(x, y, 'o-', label=label)
    plt.xlabel('h')
    plt.ylabel('error')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    # Read data from CSV files
    data_dir = "../../Cpp/compmath/export_data/"
    
    x, y_func = read_csv(data_dir + "function.csv")
    x, y_der = read_csv(data_dir + "derivative.csv")
    x, y_sec_der = read_csv(data_dir + "sec_der.csv")
    x, y_true_der = read_csv(data_dir + "true_der.csv")
    x, y_true_sec_der = read_csv(data_dir + "true_sec_der.csv")

    # Plot the functions
    plot_functions(
        x, [y_func, y_true_der, y_true_sec_der], 
        labels=["Function", "Derivative", "Second Derivative"], 
        title="Function and Its Derivative"
    )
```

```python
h_vals, errors_der_1_ord = read_csv(data_dir + "errors_der_1_ord.csv")
h_vals, errors_der_2_ord = read_csv(data_dir + "errors_der_2_ord.csv")
h_vals, errors_der_4_ord = read_csv(data_dir + "errors_der_4_ord.csv")
h_vals, errors_sec_der_2_ord = read_csv(data_dir + "errors_sec_der_2_ord.csv")
```

```python
plot_loglog_functions(
        h_vals, [errors_der_1_ord, errors_der_2_ord, errors_der_4_ord, errors_sec_der_2_ord], 
        labels=["Derivative 1 order", "Derivative 2 order", "Derivative 4 order", "Second Derivative 2 order"], 
        title="Order of approximation"
    )
```
