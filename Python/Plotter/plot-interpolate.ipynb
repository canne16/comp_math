```python
import csv
import matplotlib.pyplot as plt

def read_csv(filename):
    x = []
    y = []
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header row
        for row in reader:
            x.append(float(row[0]))
            y.append(float(row[1]))
    return x, y

def plot_functions(xs, ys, labels, title):
    plt.figure(figsize=(10, 6))

    plt.plot(xs[0], ys[0], label=labels[1], linestyle='', marker='o')
    plt.plot(xs[1], ys[1], label=labels[1])
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Read data from CSV files
data_dir = "../../Cpp/compmath/export_data/"

x, y_init = read_csv(data_dir + "initial_data.csv")
x_lagr, y_lagr = read_csv(data_dir + "lagrange.csv")
x_newt, y_newt = read_csv(data_dir + "newton.csv")

# Plot the functions
plot_functions(
    [x, x_lagr], 
    [y_init, y_lagr], 
    labels=["Function", "Lagrange", "Newton"], 
    title="Function and Its Interpolations"
)
```
